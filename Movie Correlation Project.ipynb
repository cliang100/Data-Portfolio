```python
# Importing Libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)


df = pd.read_csv(r"C:\Users\clian\Data Analyst Portfolio\Datasets\Movies\movies.csv")
```

```python
# Viewing the data

df.head()
```

```python
# Finding any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for each column

df.dtypes
```

```python
# Changing the data types of columns

df['budget'] = df['budget'].fillna(0).astype('int64')

df['gross'] = df['gross'].fillna(0).astype('int64')

df['votes'] = df['votes'].fillna(0).astype('int64')

df['runtime'] = df['runtime'].fillna(0).astype('int64')
```

```python
df
```

```python
# Create correct year column (based off released column)

df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df.drop_duplicates().head()
```

```python
# Budget high correlation
# Company high correlation
```

```python
# Making a scatterplot to test correlation hypotheses
# Budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')

plt.ylabel('Film Budget')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
# Assessing the correlation
```

```python
df.select_dtypes(include='number').corr(method='pearson') #pearson, kendall, spearman
```

```python
# Hypothesis was correct, budget and gross have high correlation
```

```python
correlation_matrix = df.select_dtypes(include='number').corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# Assessing Company correlation

df.head()
```

```python
df_quantified = df

for col_name in df_quantified.columns:
    if (df_quantified[col_name].dtype == 'object'):
        df_quantified[col_name] = df_quantified[col_name].astype('category')
        df_quantified[col_name] = df_quantified[col_name].cat.codes
        
df_quantified.head()
```

```python
correlation_matrix = df_quantified.select_dtypes(include='number').corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_quantified.corr()
```

```python
correlation_mat = df_quantified.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[sorted_pairs > 0.5]

high_corr
```
